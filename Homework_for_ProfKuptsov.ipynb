```
#3. Linear algebra, lesson 1, exercise 4

def draw_vector(ax, label, color, vec, orig):
    """
    Draw a vector
    -------------
    ax : axes
        The axes to draw
    label : str
        The label of the vector
    color : str
        Color specification
    vec : list of two components
        The vector - shifts dx and dy from orig
    orig : list of two components
        The vector beginning point
    """
    
    vec_end = [orig[0] + vec[0], orig[1] + vec[1]] 
    ax.annotate("", 
                xytext=orig,  # arrow begining point
                xy=vec_end,   # arrow end point
                arrowprops={'color': color});

    # Draw a transparent circle around a label to improve visibility
    bbox = {'facecolor':'w',
            'edgecolor': 'w',
            'boxstyle': 'circle',
            'alpha': 0.8}

    # Add a text in the middle of the arrow
    ax.text(orig[0] + 0.5*vec[0], orig[1] + 0.5*vec[1], label,
           fontsize=16, bbox=bbox, va='center', ha='center')
    
vec1 = [0.98480775,0.17364818]
vec2 = [-0.64278761,0.76604444]
vec3 = [-0.34202014,-0.93969262]

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(5,5))

ax = plt.gca()
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])

draw_vector(ax, label=r'$\vec1 v$', color='C1', vec=vec1, orig=[0,0])
draw_vector(ax, label=r'$\vec2 v$', color='C2', vec=vec2, orig=[0,0])
draw_vector(ax, label=r'$\vec3 v$', color='C3', vec=vec3, orig=[0,0])

ax.set_xlabel('$x_0$')
ax.set_ylabel('$x_1$')

ax.grid();
```

```
#4. Statistics, lesson 1, ex.1

import requests
base_url = "https://raw.githubusercontent.com/kupav/data-sc-intro/main/data/"
file_name = "bank_customers.csv"
web_data = requests.get(base_url + file_name)
assert web_data.status_code == 200
print(web_data.text[:100])
str_data = web_data.text.splitlines()
print(str_data[:10])
lst_data = [s.split(',') for s in str_data[1:]]
print(lst_data[:10])
data_months = [int(s[1]) for s in lst_data]
print(data_months[:10])

def my_mean(data_months):
    return sum(data_months)/len(data_months)

data_mean=my_mean(data_months)

def my_median(data_months):
    size = len(data_months)
    srt_data = sorted(data_months)
    if size % 2 != 0:
        midpoint = size // 2
        return srt_data[midpoint]
    else:
        hi_midpoint = size // 2
        lo_midpoint = hi_midpoint - 1
        return (srt_data[lo_midpoint] + srt_data[hi_midpoint])/2

data_median=my_median(data_months)

from collections import Counter

def my_mode(data_months):
    bins = Counter(data_months)
    most_freq = bins.most_common(1)[0][1]
    return [key for key, val in bins.items() if val == most_freq]

data_mode=my_mode(data_months)

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.hist(data_months, bins=max(data_months) - min(data_months) + 1);
ax.set_xlabel("# months in relationships")
ax.set_ylabel("# of customers")
ax.axvline(data_mean, color='red', linewidth=3, label=f'mean at {data_mean:6.2f}')
ax.axvline(data_median, color='green', linewidth=3, label=f'median at {data_median}')
for md in data_mode:
    ax.axvline(md, color='cyan', linewidth=3, label=f'mode at {md}')
ax.legend()
ax.grid();
```

```
#6. Hypothesis and inference. Lesson 1. Ex.1
import numpy as np
from scipy.special import erf

mu, sig = 0, 1
lo, hi = -1, 1

def norm_cdf(x, mu, sig):
    return 0.5 * (1 + erf((x-mu)/(sig*np.sqrt(2))))

def norm_btw(lo, hi, mu, sig):
    assert lo < hi
    p1 = norm_cdf(lo, mu, sig)
    p2 = norm_cdf(hi, mu, sig)
    return p2 - p1
print (norm_btw(lo, hi, mu, sig))
```

**11. Basic concepts of machine learning. Lesson 1, ex.1.**

Model is specification of mathematical (or probabilistic) relationships between different variables, in other words, it is a formalized description of relationships. They must include the most important properties and relations of another system. 

Machine learning model is a program that has been trained to recognize certain types of patterns. You train a model over a set of data, providing it an algorithm that it can use to reason over and learn from those data.

**11. Basic concepts of machine learning. Lesson 1, ex.2.**
Supervised models are trained with the help of labeled dataset (it means that somebody worked it through in advance and added all the desired output). This kind of model will compare the produced output with the desired one and in the case of finding any difference the difference will be eliminated.

Unsupervised models are also trained on a dataset, but it doesn't have any desired outputs. The purpose of using this kind of learning is to see patterns and groups (clusters) within the data.

Reinforcement models are close to supervised, but they don't have a desired output as they are used for changing, dynamic sutuations.


**11. Basic concepts of machine learning. Lesson 2, ex.1.**

Overfitting means that the learning model is far too dependent on training data. In other words, it demonstrates good performance on the training data and poor generliazation to other data.
Underfitting means that the model has a poor relationship with the training data. In other words, it demonstrates poor performance on the training data and poor generalization to other data.
Dataset must be splitted into three parts as they are going to be used as a basis for training, validation and test. Training data are immediately used to update model parameters. Validation data are used after several steps of training to estimate the performance and overfitting of the model; on the basis of this estimation corrections to training can be done. Test data are used just once to get the final score of the model; no model parameter updates can be done after that.

**11. Basic concepts of machine learning. Lesson 2, ex.5.**
Cross-validation is a method of working with small dataset. A dataset should be splitted into k parts, and every k-1 part is used for the model training and test the model using the last one. **16. Vector models of text. Lesson 1. Ex.2**
Stemming is a process of removing the last few characters of a word(which are actually the inflectional forms of a word) in order to come to its initial form(stem).

Lemmatization is a process of finding lemma (a meaningful base form of a word) through analysing a context and further converting of the initial word to a lemma itself.

As stemming is less accurate way of receiving a root of a word, there are high chances to receive invalid word through using stemming, while using lemmatization have higher chances to finish successfully.

Speaking about the purposes of using stemming and lemmatization, those means are widely used in tagging systems, indexing, SEOs, Web search results, and information retrieval.

The Bag of Words (BoW) model is the simplest form of text representation in numbers. Like the term itself, we can represent a sentence as a bag of words vector (a string of numbers). In Bag of words, one can extract only the unigram words to create unordered list of words without syntactic, semantic and POS tagging. This bunch of words represent the document.

TF-IDF is a word-document mapping (with some normalization). It ignores the order of words and gives nxm matrix (or mxn depending on implementation) where n is number of words in the vocabulary and m is number of documents. Tf-idf is a scoring scheme for words - that is a measure of how important a word is to a document.
